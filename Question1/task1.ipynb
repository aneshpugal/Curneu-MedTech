```python
import csv
#regular expression
import re
from os import listdir
#used to list directory files
list_of_files = listdir('X:stuff/Curneu MedTech/collaborativeData/')

#special case for my approach in dealing with 'Recommended' and 'Selected' columns identified by reading
#all unique values of each file

special_case=['Neck-Futu(LI18)','Neck-FutuLI18']

#iteration counter
iter_counter=0

#file counter used for iterating files dynamically
file_counter=1

#iterating through all the files
for i in list_of_files:
    
    #Creating dynamically file in the specified path by iterating using file counter
    with open("X:stuff/Curneu MedTech/user%d.csv" % (file_counter),"w",newline="") as f:
        
        #Opening dynamically file in the specified path by iterating using iteration counter
        with open('X:stuff/Curneu MedTech/collaborativeData/%s' % (list_of_files[iter_counter]), "r") as csv_file:
            
            #file is read into csv_reader
            csv_reader = csv.reader(csv_file, delimiter=',')
            
            #storing first row of data(header) of each file in lines variable as a list
            lines=next(csv_reader,None)
            
            #the list item is stored into a dataframe 
            df=[[lines[0],lines[1],lines[2],lines[3],lines[4],lines[5]]]
            
            #mapping to the file to be written
            writer = csv.writer(f)
            
            #writing the dataframe as a single row to the newly created file 
            writer.writerows(df)
            
            #iterating through the list consisting of one row of data
            for lines in csv_reader:
                
                #First approach using regular expression for filtering 'Selected' column of the data
                gf=re.findall('[A-Z]*[a-z]*[0-9]*', lines[5])
                
                #Second approach for filtering 'Recommended Column'
                #APPROACH - Append data starting from second capital letter using position_counter and capital_counter
                #capital letter counter
                capital_counter=0
                
                #position counter for slicing the word  based on position
                pos=0
                
                #storing 'Recommended' column data 
                strty=lines[4]
               
                #checking if the word is in the special case list if so the capital_counter is decremented
                if strty in special_case:
                    capital_counter=-1
                
                #ierating through the word for accessing individual character for count of capital letters
                for q in strty:
                    
                    #method to check if the character is capital letter if so increment
                    if q.isupper():
                        capital_counter=capital_counter+1
                    #increment the position counter for keeping track of the second capital letter
                    pos=pos+1
                    
                    #if second capital letter is found we break the ieration 
                    if (capital_counter==2):
                        break
                
                #I check for the file iteration counter since in first file 'Recommended' column has data different to that of other files
                if iter_counter==0:
                    
                    #appending the first three columns as such to the dataframe and the fourth column is appended by slicing from 
                    #the position of occurence of second capital letter and neglecting last character as that's the special case
                    #in file1 as it includes paranthesis so while slicing we ignore the ')' character
                    
                    df=[[lines[0],lines[1],lines[2],lines[3],strty[(pos-1):-1],gf[1]]]
                    
                    #writing the dataframe as a single row to the newly created file 
                    writer = csv.writer(f)
                    writer.writerows(df)
                
                #if the file counter is >0 there's no special case so we modify the slicing using second approach
                if iter_counter>0:
                    
                    df=[[lines[0],lines[1],lines[2],lines[3],strty[(pos-1):],gf[1]]]
                    writer = csv.writer(f)
                    writer.writerows(df)
    
    #incrementing counters
    file_counter=file_counter+1
    iter_counter=iter_counter+1
```
